```python
# Covert 'FireDiscoveryDateTime' to datetime format
df['FireDiscoveryDateTime'] = pd.to_datetime(df['FireDiscoveryDateTime'])

# Convert to Unix timestamp
df['FireDiscoveryDateTime_unix'] = df['FireDiscoveryDateTime'].astype('int64') / 10**9

# Input data to heat_data and ensure all variables same type
heat_data = [
    [float(row["initial_latitudes"]), float(row["initial_longitudes"]), float(row["FireDiscoveryDateTime_unix"])] 
    for _, row in df.iterrows() 
]


# Create a base map for the heatmap
m = folium.Map(location=[37.0902, -95.7129], zoom_start=4)


# Hover effects, loop through the data. Code generated by OpenAI's ChatGPT (January 2025)
for _, row in df.iterrows():
    tooltip_text = f"""
    <strong>Discovered:</strong> {row['FireDiscoveryDateTime']}<br>
    <strong>location:</strong> ({row['initial_latitudes']}, {row['initial_longitudes']})
    """
    folium.CircleMarker(
        location=[row['initial_latitudes'], row['initial_longitudes']],
        radius=3,
        color="blue",
        fill=True,
        fill_opacity=0.6,
        tooltip=tooltip_text
    ).add_to(m)
```

```python
# heatmap layer
HeatMap(heat_data, radius=35, blur=15, max_zoom=1).add_to(m)
```
